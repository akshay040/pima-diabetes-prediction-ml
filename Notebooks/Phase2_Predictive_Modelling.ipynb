# <u>Predictive Modelling</u>

<a id="Feature Selection"></a>
### <u>Feature Selection</u>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, RepeatedStratifiedKFold
from sklearn.metrics import classification_report, f1_score, precision_score, recall_score, confusion_matrix
from sklearn.feature_selection import SelectKBest, f_classif

# Load the dataset
data = pd.read_csv('diabetes.csv')

# Data Cleaning and Preprocessing
data = data.drop_duplicates()
data['Glucose'] = data['Glucose'].replace(0, data['Glucose'].mean())
data['BloodPressure'] = data['BloodPressure'].replace(0, data['BloodPressure'].mean())
data['SkinThickness'] = data['SkinThickness'].replace(0, data['SkinThickness'].median())
data['Insulin'] = data['Insulin'].replace(0, data['Insulin'].median())
data['BMI'] = data['BMI'].replace(0, data['BMI'].median())

# Removing outliers based on provided criteria
data = data[data["Pregnancies"] <= 14]
data = data[data["Glucose"] >= 50]
data = data[data["BloodPressure"] <= 120]
data = data[data["SkinThickness"] <= 80]
data = data[data["Insulin"] <= 600]
data = data[data["BMI"] <= 55]
data = data[data["DiabetesPedigreeFunction"] <= 2]

# Splitting the Data Frame into X and y
target_name = 'Outcome'
y = data[target_name]
X = data.drop(target_name, axis=1)

# Feature Selection
selector = SelectKBest(score_func=f_classif, k='all')
X_new = selector.fit_transform(X, y)

# Get scores
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': selector.scores_})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

# Plotting feature scores
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Feature', data=feature_scores)
plt.title('Feature Importance Scores')
plt.show()

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<a id="Model Fitting & Tuning"></a>
# <u>Model Fitting & Tuning</u>

<a id="K-Nearest Neighbors (KNN)"></a>
### <u>K-Nearest Neighbors (KNN)</u>

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
param_grid_knn = {
    'n_neighbors': list(range(15, 25)),
    'p': [1, 2],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search_knn = GridSearchCV(knn, param_grid_knn, cv=cv, scoring='f1', n_jobs=-1, error_score=0)
best_model_knn = grid_search_knn.fit(X_train, y_train)

print("Best hyperparameters for KNN:", best_model_knn.best_params_)

knn_pred = best_model_knn.predict(X_test)
print("KNN Classification Report:\n", classification_report(y_test, knn_pred))
print("KNN F1 Score:\n", f1_score(y_test, knn_pred))
print("KNN Precision Score:\n", precision_score(y_test, knn_pred))
print("KNN Recall Score:\n", recall_score(y_test, knn_pred))
sns.heatmap(confusion_matrix(y_test, knn_pred), annot=True)
plt.title("KNN Confusion Matrix")
plt.show()
```

## <u>Neural Network Model Fitting & Tuning</u>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

# Initialize parameters
def initialize_parameters(input_size, hidden_size1, hidden_size2, output_size):
    np.random.seed(1)
    w_input_hidden1 = np.random.randn(hidden_size1, input_size) * 0.01
    b_hidden1 = np.zeros((hidden_size1, 1))
    w_hidden1_hidden2 = np.random.randn(hidden_size2, hidden_size1) * 0.01
    b_hidden2 = np.zeros((hidden_size2, 1))
    w_hidden2_output = np.random.randn(output_size, hidden_size2) * 0.01
    b_output = np.zeros((output_size, 1))
    return w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output

# Activation functions
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu_derivative(z):
    return z > 0

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

# Dropout function
def dropout(a, keep_prob):
    mask = np.random.rand(*a.shape) < keep_prob
    return a * mask / keep_prob

# Forward propagation
def forward_propagation(X, w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output, keep_prob=1.0):
    z_hidden1 = np.dot(w_input_hidden1, X) + b_hidden1
    a_hidden1 = relu(z_hidden1)
    a_hidden1 = dropout(a_hidden1, keep_prob)
    
    z_hidden2 = np.dot(w_hidden1_hidden2, a_hidden1) + b_hidden2
    a_hidden2 = relu(z_hidden2)
    a_hidden2 = dropout(a_hidden2, keep_prob)
    
    z_output = np.dot(w_hidden2_output, a_hidden2) + b_output
    a_output = sigmoid(z_output)
    
    return a_hidden1, a_hidden2, a_output

# Compute loss
def compute_loss(y, a_output):
    m = y.shape[1]
    loss = -np.sum(y * np.log(a_output) + (1 - y) * np.log(1 - a_output)) / m
    return loss

# Compute accuracy
def compute_accuracy(y, a_output):
    predictions = (a_output > 0.5)
    accuracy = np.mean(predictions == y)
    return accuracy

# Backpropagation
def backpropagation(X, y, a_hidden1, a_hidden2, a_output, w_input_hidden1, w_hidden1_hidden2, w_hidden2_output, keep_prob=1.0):
    m = X.shape[1]

    dz_output = a_output - y
    dw_hidden2_output = (1 / m) * np.dot(dz_output, a_hidden2.T)
    db_output = (1 / m) * np.sum(dz_output, axis=1, keepdims=True)

    dz_hidden2 = np.dot(w_hidden2_output.T, dz_output) * relu_derivative(a_hidden2)
    dz_hidden2 = dropout(dz_hidden2, keep_prob)
    dw_hidden1_hidden2 = (1 / m) * np.dot(dz_hidden2, a_hidden1.T)
    db_hidden2 = (1 / m) * np.sum(dz_hidden2, axis=1, keepdims=True)

    dz_hidden1 = np.dot(w_hidden1_hidden2.T, dz_hidden2) * relu_derivative(a_hidden1)
    dz_hidden1 = dropout(dz_hidden1, keep_prob)
    dw_input_hidden1 = (1 / m) * np.dot(dz_hidden1, X.T)
    db_hidden1 = (1 / m) * np.sum(dz_hidden1, axis=1, keepdims=True)

    return dw_input_hidden1, db_hidden1, dw_hidden1_hidden2, db_hidden2, dw_hidden2_output, db_output

# Update parameters
def update_parameters(w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output, 
                      dw_input_hidden1, db_hidden1, dw_hidden1_hidden2, db_hidden2, dw_hidden2_output, db_output, learning_rate):
    w_input_hidden1 -= learning_rate * dw_input_hidden1
    b_hidden1 -= learning_rate * db_hidden1
    w_hidden1_hidden2 -= learning_rate * dw_hidden1_hidden2
    b_hidden2 -= learning_rate * db_hidden2
    w_hidden2_output -= learning_rate * dw_hidden2_output
    b_output -= learning_rate * db_output
    return w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output

# Training Loop
input_size = X_train.shape[1]
hidden_size1 = 12
hidden_size2 = 6
output_size = 1

# Ensure y_train and y_test are NumPy arrays
y_train_array = y_train.values.flatten()
y_test_array = y_test.values.flatten()

# Reshape y_train and y_test to 2D arrays with one row
y_train_array = y_train_array.reshape(1, -1)
y_test_array = y_test_array.reshape(1, -1)

# Initialize parameters
w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output = initialize_parameters(input_size, hidden_size1, hidden_size2, output_size)

# Hyperparameters
learning_rate = 0.001
num_epochs = 50
keep_prob = 0.8

# Lists to store accuracy, loss
accuracy_list = []
loss_list = []

for epoch in range(num_epochs):
    # Forward propagation
    a_hidden1, a_hidden2, a_output = forward_propagation(X_train.T, w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output, keep_prob)
    
    # Calculate loss
    loss = compute_loss(y_train_array, a_output)
    
    # Calculate accuracy
    accuracy = compute_accuracy(y_train_array, a_output)
    
    # Backpropagation
    dw_input_hidden1, db_hidden1, dw_hidden1_hidden2, db_hidden2, dw_hidden2_output, db_output = backpropagation(X_train.T, y_train_array, a_hidden1, a_hidden2, a_output, w_input_hidden1, w_hidden1_hidden2, w_hidden2_output, keep_prob)
    
    # Update parameters
    w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output = update_parameters(w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output, 
                                                                                                            dw_input_hidden1, db_hidden1, dw_hidden1_hidden2, db_hidden2, dw_hidden2_output, db_output, learning_rate)
    
    # Print accuracy and loss 
    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch + 1}/{num_epochs} - accuracy: {accuracy:.4f} - loss: {loss:.4f}")
    
    # Save accuracy and loss values
    accuracy_list.append(accuracy)
    loss_list.append(loss)

# Plotting Loss and Accuracy
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6))
fig.subplots_adjust(wspace=0.3)

# Plot Loss
ax1.plot(loss_list, color='tab:blue')
ax1.set_title("Loss Plot")
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Plot Accuracy
ax2.plot(accuracy_list, color='tab:orange')
ax2.set_title("Accuracy Plot")
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

# Add a suptitle
plt.suptitle("Training Progress", fontsize=16)

# Show the plots
plt.tight_layout()
plt.show()

# Evaluating the model
def evaluate(y_true, y_preds):
    cf = confusion_matrix(y_true, y_preds)
    TP = cf[0][0]
    FP = cf[0][1]
    FN = cf[1][0]
    TN = cf[1][1]
    
    TPR = TP / (TP + FN) if (TP + FN) > 0 else 0
    FPR = FP / (FP + TN) if (FP + TN) > 0 else 0
    
    print(f'True Positive Rate: {round(TPR, 4)}')
    print(f'False Positive Rate: {round(FPR, 4)}')

# Predictions
def get_predictions(probs):
    return (probs > 0.5).astype(int).flatten()

def make_predictions(X_test, w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output, keep_prob=1.0):
    _, _, a_output = forward_propagation(X_test.T, w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output, keep_prob)
    predictions = get_predictions(a_output)
    return predictions

y_train_preds = make_predictions(X_train, w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output)
print('Model Evaluation, Train Dataset')
evaluate(y_train_array.flatten(), y_train_preds)

y_test_preds = make_predictions(X_test, w_input_hidden1, b_hidden1, w_hidden1_hidden2, b_hidden2, w_hidden2_output, b_output)
print('Model Evaluation, Test Dataset')
evaluate(y_test_array.flatten(), y_test_preds)
```

### <u>Discussion</u>

#### Overview

In this task, we built a neural network model from scratch using Python and NumPy to perform binary classification. We then fine-tuned the model by adjusting various hyperparameters to optimize its performance. This discussion provides an accurate, detailed, and insightful overview of the neural network model, including its topology and all relevant model parameter values used. Additionally, it presents the fine-tuning of at least five different neural network hyperparameters along with proper narratives and sufficient explanation.

#### Neural Network Topology

Our neural network consists of the following layers:

1. **Input Layer**: The input layer's size is determined by the number of features in the dataset. For our specific task, the input size is determined by `X_train.shape[1]`.
2. **First Hidden Layer**: This layer consists of 12 neurons and uses the Rectified Linear Unit (ReLU) activation function. The ReLU function introduces non-linearity to the model, which helps it learn complex patterns in the data.
3. **Second Hidden Layer**: This layer consists of 6 neurons and also uses the ReLU activation function. The addition of this second hidden layer helps the model capture even more complex relationships within the data.
4. **Output Layer**: This layer has 1 neuron and uses the Sigmoid activation function. The sigmoid function maps the output to a probability between 0 and 1, which is suitable for binary classification tasks.

The neural network's parameters, including weights and biases, are initialized as follows:
- **Weights**: Initialized using a normal distribution scaled by 0.01 to ensure small initial values, preventing issues like vanishing or exploding gradients.
- **Biases**: Initialized to zeros.

#### Activation Functions

Two activation functions are used in our network:
- **ReLU (Rectified Linear Unit)**: Applied in the hidden layers to introduce non-linearity. The ReLU function sets negative values to zero and keeps positive values unchanged.
- **Sigmoid**: Applied in the output layer to map the output to a probability between 0 and 1, which is essential for binary classification tasks.

#### Forward Propagation

During forward propagation, input data is passed through the layers, applying weights, biases, and activation functions. Additionally, dropout is applied to the hidden layers to prevent overfitting. Dropout works by randomly setting some activations to zero during training, which forces the network to learn more robust features that are not reliant on any specific neurons.

#### Backpropagation

In backpropagation, gradients of the loss with respect to the parameters are calculated using the chain rule. These gradients guide the parameter updates in the direction that minimizes the loss. The steps involved in backpropagation are as follows:
- **Calculate the gradients for the output layer**.
- **Calculate the gradients for the second hidden layer**.
- **Calculate the gradients for the first hidden layer**.
- **Use the gradients to update the weights and biases**.

#### Parameter Updates

Parameters are updated using gradient descent. The learning rate controls the step size of the updates. In our model, the learning rate is set to 0.001.

#### Loss and Accuracy Calculation

The loss is computed using binary cross-entropy, which quantifies the difference between the predicted and actual labels. The formula for binary cross-entropy is:

\[ L = -\frac{1}{n} \sum [y_i \log(a_i) + (1 - y_i) \log(1 - a_i)] \]

Where \( y_i \) is the actual label and \( a_i \) is the predicted probability. 

Accuracy is computed as the proportion of correct predictions, providing a measure of how well the model performs on the training data.

#### Fine-Tuning Hyperparameters

We fine-tuned the following hyperparameters to improve the model's performance:

1. **Learning Rate**: Determines the step size during gradient descent. We tested learning rates of 0.001 and 0.005.
2. **Number of Hidden Nodes**: Determines the model's capacity to learn patterns. We experimented with different sizes for the hidden layers.
3. **Dropout Probability**: Prevents overfitting by randomly dropping units during training. We tested dropout probabilities of 0.5 and 0.8.

The combinations of hyperparameters tested are as follows:

```python
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.metrics import roc_auc_score
from scipy.stats import ttest_rel

# Updated KerasClassifierWrapper with classes_
class KerasClassifierWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, optimizer='adam', init='uniform', batch_size=10, epochs=10):
        self.optimizer = optimizer
        self.init = init
        self.batch_size = batch_size
        self.epochs = epochs
        self.model = None
        self.classes_ = None
    
    def build_model(self):
        model = Sequential()
        model.add(Input(shape=(X_train.shape[1],)))
        model.add(Dense(12, kernel_initializer=self.init, activation='relu'))
        model.add(Dense(8, kernel_initializer=self.init, activation='relu'))
        model.add(Dense(1, kernel_initializer=self.init, activation='sigmoid'))
        model.compile(loss='binary_crossentropy', optimizer=self.optimizer, metrics=['accuracy'])
        return model
    
    def fit(self, X, y):
        self.classes_ = np.unique(y)
        self.model = self.build_model()
        self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, verbose=0)
        return self
    
    def predict(self, X):
        return (self.model.predict(X) > 0.5).astype("int32")
    
    def predict_proba(self, X):
        proba = self.model.predict(X)
        return np.hstack([1-proba, proba])  # Return probabilities for both classes

# Initialize dictionary to store AUC scores from each fold
auc_scores = {
    'Logistic Regression': [],
    'Decision Tree': [],
    'Random Forest': [],
    'Gradient Boosting': [],
    'Neural Network': []
}

# Number of folds for cross-validation
cv = StratifiedKFold(n_splits=5)

# Function to evaluate and store AUC scores
def evaluate_model(model, model_name, X, y):
    aucs = []
    for train_idx, test_idx in cv.split(X, y):
        X_train_cv, X_test_cv = X.iloc[train_idx], X.iloc[test_idx]
        y_train_cv, y_test_cv = y.iloc[train_idx], y.iloc[test_idx]
        
        model.fit(X_train_cv, y_train_cv)
        y_pred_proba = model.predict_proba(X_test_cv)[:, 1]
        auc = roc_auc_score(y_test_cv, y_pred_proba)
        aucs.append(auc)
    
    auc_scores[model_name].extend(aucs)

# Logistic Regression
logreg = LogisticRegression(max_iter=1000)  # Increased max_iter to help with convergence
evaluate_model(logreg, 'Logistic Regression', X_train, y_train)

# Decision Tree
dtree = DecisionTreeClassifier()
evaluate_model(dtree, 'Decision Tree', X_train, y_train)

# Random Forest
rf = RandomForestClassifier()
evaluate_model(rf, 'Random Forest', X_train, y_train)

# Gradient Boosting
gb = GradientBoostingClassifier()
evaluate_model(gb, 'Gradient Boosting', X_train, y_train)

# Neural Network
nn = KerasClassifierWrapper(optimizer='adam', init='uniform', batch_size=10, epochs=10)
evaluate_model(nn, 'Neural Network', X_train, y_train)

# Print AUC scores
print("AUC Scores:")
for model, scores in auc_scores.items():
    print(f"{model}: {np.mean(scores):.2f}")

# Performing paired t-tests
print("\nPaired t-tests between models' AUC scores:")
models = list(auc_scores.keys())
for i in range(len(models)):
    for j in range(i + 1, len(models)):
        model1, model2 = models[i], models[j]
        t_stat, p_val = ttest_rel(auc_scores[model1], auc_scores[model2])
        print(f"{model1} vs {model2}: t-statistic = {t_stat:.2f}, p-value = {p_val:.2f}")
```

Conclusion:
Logistic Regression demonstrates strong performance, significantly outperforming Decision Tree and Neural Network.
Random Forest and Gradient Boosting both perform well, with no significant difference between them.
Decision Tree and Neural Network have the lowest AUC scores and are outperformed by other models in most comparisons.
The paired t-tests provide a robust statistical basis for comparing the models' performances, indicating where significant differences exist.
